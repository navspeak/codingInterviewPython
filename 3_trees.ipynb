```python
class TreeNode:
    def __init__(self, val, left = None, right = None):
        self.value = val
        self.left = left
        self.right = right

four = TreeNode(4)
five = TreeNode(5)
two = TreeNode(2, four, five)
three = TreeNode(3)
one = TreeNode(1, two, three)

    #     1
    #    / \
    #   2   3
    #  / \
    # 4   5


def preorder(head):
    if not head:
        return
    print(head.value, end=", ")
    preorder(head.left)
    preorder(head.right)

#Inorder of a BST is always sorted
def inorder(head):
    if not head:
        return
    inorder(head.left)
    print(head.value, end=", ")
    inorder(head.right)

def postorder(head):
    if not head:
        return
    postorder(head.left)
    postorder(head.right)
    print(head.value, end=", ")

preorder(one)
print("")

inorder(one)
print("")

postorder(one)
print("")
```

```python
def inorder_v2(head):
    if not head: return []
    return inorder_v2(head.left) + [head.value] + inorder_v2(head.right)

print(inorder_v2(one))
```

```python
def preorder_stack(head):
    stack = [head]
    ret = []
    while stack:
        curr = stack.pop()
        ret.append(curr.value)
        # Stack is LIFO
        # Preorder = root → left → right
        # So we must push right first, then left
        if curr.right: stack.append(curr.right)
        if curr.left: stack.append(curr.left)
    return ret

print(preorder_stack(one))
print(preorder(one))
```

```python
def inorder_stack(head):
    stack = [head]
    ret = []
    curr = head
    while curr or stack:
        while curr:
            stack.append(curr)
            curr = curr.left
        curr = stack.pop()
        ret.append(curr.value)
        curr = curr.right
    return ret

print(inorder_stack(one))
print(inorder(one))
postorder(one)
```

```python
# Given: (note both pre and inorder are needed)
# preorder = [1, 2, 4, 5, 3]
# inorder = [4, 2, 5, 1, 3]
# Reconstruct the same tree:
#         1
#        / \
#       2   3
#      / \
#     4   5

def reconstructBT(preorder, inorder):
    if not preorder or not inorder: return None

    head = TreeNode(preorder[0])
    mid = inorder.index(preorder[0])
    head.left = reconstructBT(preorder[1:mid+1], inorder[:mid])
    head.right = reconstructBT(preorder[mid+1:], inorder[mid+1:])
    return head
    #  [1, (2, 4, 5,| 3)]
    #  [(4, 2, 5,) 1, 3]  => mid = 3=> L will have 3 or mid number of elems []
    # -------------------
    #   0  1  2   3   4

    # index is O(n), slicing too => n^2 => also index to work assumes unique
    # With duplicate values, preorder + inorder does NOT uniquely identify a binary tree unless you add extra information.
```
