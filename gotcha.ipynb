```python
def custom_sort(lst):
    def key_fn(x):
        if isinstance(x, str) and x.isdigit():
            return (0, int(x))          # numeric strings
        elif isinstance(x, str):
            return (1, x)               # other strings (lexical)
        else:
            return (2, 0)               # keep relative order

    return sorted(lst, key=key_fn)

data = ["10", [1,2], "2", "apple", ["x"], "banana", "1"]

print(custom_sort(data))
```

```python
import bisect

arr = [1, 3, 5, 7]

i = bisect.bisect_left(arr, 5)
print(i)   # 2
i = bisect.bisect(arr, 5)
print(i)   # 3

#inserting while keeping list sorted
arr = [1, 3, 5, 5, 5, 7]
bisect.insort_right(arr, 4)
print(arr)   # [1, 3, 4, 5, 5, 5, 7]
```

#### OOPs
- supports multiple inheritance
- MRO = Method Resolution Order (MRO) = the order in which Python looks for a method when you call it. (super = next in MRO, not parent)

```python
class A:
    def foo(self): print("A foo")

class B:
    def foo(self): print("B foo")
    def bar(self): print("B bar")

class C(B, A):
    pass

c = C()
c.foo()
c.bar()
c.foo()
print(C.mro())
```

```python
class A:
    def foo(self):
        print("A.foo")

class B(A):
    def foo(self):
        print("B.foo")
        super().foo()

class C(A):
    def foo(self):
        print("C.foo")
        super().foo()

class D(B, C):
    pass

D().foo()
# Note: Because super() means: call next method in MRO, not parent class
# B.foo
# C.foo
# A.foo 

# and not
# B.foo
# A.foo
```

```python
from abc import ABC, abstractmethod

class PaymentProcessor(ABC):

    @abstractmethod
    def pay(self, amount: float):
        pass

class Stripe(PaymentProcessor):
    def pay(self, amount):
        print("Paid", amount)

Stripe()  # OK
PaymentProcessor()  # ❌ TypeError
```
