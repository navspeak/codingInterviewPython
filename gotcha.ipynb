```python
#https://leetcode.com/problems/longest-repeating-character-replacement/description/
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        left = maxlen = 0
        char_freq = defaultdict(int)

        for right, alphabet in enumerate(s):
            char_freq[alphabet] +=1
            window_size = right - left + 1
            num_of_replacments_needed = window_size - max(char_freq.values()) # this is O(n) => let's optimize
            while num_of_replacments_needed > k:
                char_freq[s[left]] -=1
                left +=1
                window_size = right - left + 1
                num_of_replacments_needed = window_size - max(char_freq.values())
            maxlen = max(maxlen, window_size)
        return maxlen

    def characterReplacement_op(self, s: str, k: int) -> int:
        left = maxlen = 0
        char_freq = defaultdict(int)
        max_freq = 0
        for right, alphabet in enumerate(s):
            char_freq[alphabet] +=1
            window_size = right - left + 1
            max_freq = max(max_freq, char_freq[alphabet])
            num_of_replacments_needed = window_size - max_freq # this is O(n) => let's optimize
            while num_of_replacments_needed > k:
                char_freq[s[left]] -=1
                left +=1
                window_size = right - left + 1
                num_of_replacments_needed = window_size - max_freq
            maxlen = max(maxlen, window_size)
        return maxlen

```
