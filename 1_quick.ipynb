# LIST

### Add 

| Method            | Example                   | Result               | Notes
|-|-|-|-|
| append(x)         | nums.append(4)            | [1,2,3,4]            | |
| extend(iterable)  | nums.extend([5,6])        | [1,2,3,4,5,6]        | |
| insert(i, x)      | nums.insert(1, 99)        | [1,99,2,3]           | |
| + operator        | nums + [7,8]              | new list created     | |

### Remove
| Method            | Example                   | Notes                | notes |
|-|-|-|-|
| remove(x)         | nums.remove(3)            | Removes first match  |  no new list. In place modification. Raises ValueError|
| pop()             | nums.pop()                | Removes last         | |
| pop(i)            | nums.pop(0)               | Removes by index     | |
| del nums[index]   | del nums[1]               | No return value      | no new list. In place modification. Raises IndexError |
| clear()           | nums.clear()              | Empties list         | |



| Operation         | Example                   |
|-|-|
| Access element    | nums[0], nums[-1]         |
| Length            | len(nums)                 |
| Membership        | 3 in nums                 |
| Loop              | for x in nums:            |



```python
# +------------------+------------------+------------------+---------------------+
# | Method           | In-place clear   | Preserves refs   | Idiomatic           |
# +------------------+------------------+------------------+---------------------+
# | nums.clear()     | Yes              | Yes              | ⭐ Best             |
# | del nums[:]      | Yes              | Yes              | OK (old-school)     |
# | nums[:] = []     | Yes              | Yes              | OK (less readable)  |
# | nums = []        | No               | No               | ❌ Rebinds name     |
# +------------------+------------------+------------------+---------------------+
```
