```python
class Solution(object):
    #https://leetcode.com/problems/longest-palindromic-substring/ => O(n^2)
    def longestPalindrome(self, s):
        """
        :type s: str
        :rtype: str
        """
        def expand_from_middle(s, left, right):
            l, r = left, right
            while l >= 0 and r < len(s) and s[l] == s[r]:
                l = l - 1
                r = r + 1
            return r - (l+1)

        start, end = 0, 0
        maxlen = 1
        for i in range(len(s)):
            len1 = expand_from_middle(s, i, i)
            len2 = expand_from_middle(s, i, i+1)
            newlen = max(len1, len2)
            if (newlen > maxlen):
                start = i - (newlen -1) //2
                end = i + newlen // 2
                maxlen = newlen
        return s[start:end+1]
    
```

```python
#https://leetcode.com/problems/3sum
class Solution1(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        nums = sorted(nums)
        def twoSum(nums, target, left, right):
            l, r = left, right
            result = []
            while l < r:
                if nums[l] + nums[r] == target:
                    result.append([nums[l], nums[r]])
                    l = l+1
                    r = r-1
                    while l < r and nums[l] == nums[l-1]:
                        l = l + 1
                    while r > l and nums[r] == nums[r+1]:
                        r = r - 1                  
                elif nums[l] + nums[r] > target:
                    r = r - 1
                else:
                    l = l + 1
            return result
        result = []
        for i in range(len(nums) - 2):
            if i > 0 and nums[i] == nums[i-1]: continue
            twoSumResult = twoSum(nums, -nums[i], i + 1, len(nums) - 1)
            for pair in twoSumResult:
                result.append([nums[i]] + pair)
        return result
    
print(Solution1().threeSum([-2,0,1,1,2]))
```

```python
#https://leetcode.com/problems/two-sum/description/
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        seen = {}
        result = []
        for i, num in enumerate(nums):
            rem = target - num
            if rem in seen:
                result.extend([[x, i] for x in seen[rem]])
            seen.setdefault(num,[]).append(i)
        if result:
            return result[0]
        else:
            return []
```

```python
# https://leetcode.com/problems/sort-colors/description/
class Solution(object):
    def sortColors(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        start, mid, end = 0, 0, len(nums) - 1
        while mid <= end:
            if nums[mid] == 0:
                nums[start], nums[mid] = nums[mid], nums[start]
                start+=1
                mid +=1
            elif nums[mid] == 1:
                mid+=1
            else:
                nums[end], nums[mid] = nums[mid], nums[end]
                end-=1
         
        
nums = [0, 1, 2, 2, 0, 1, 2]             
Solution().sortColors(nums)
nums

```

```python
from typing import List

def moveZerosToEnd(arr: List[int]) -> List[int]:
    left = 0
    for right in range(len(arr)):
        if arr[right]:
            arr[left], arr[right]  = arr[right], arr[left]
            left+=1
    return arr

print(moveZerosToEnd([6, 1, 0, 3, 0, -1, 2]))
```
